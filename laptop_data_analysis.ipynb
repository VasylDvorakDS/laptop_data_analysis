```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv', encoding='latin1')
df.head()
```

```python
# df['Cpu_Company'] = df['Cpu'].apply(lambda i : i.split(' ')[0])
df['Cpu_Company'] = df['Cpu'].str.split(' ').str.get(0)
df.head()
```

```python
def convert_to_gb(x):
    memory=x.split(' ')[0]
    if memory.endswith('GB'):
        return int(memory.replace('GB', ''))
    else: 
        return int(float(memory.replace('TB', ''))*1024)
```

```python
df['Memory_Amount'] = df['Memory'].apply(convert_to_gb)
#df['Memory_Amount'] = df['Memory'].apply(lambda i: int(i.split(' ')[0].replace('GB', '')) if 'GB' in i else int(float(i.split(' ')[0].replace('TB', ''))*1024))

df.head()
```

```python
def convert_to_type(x):
    
    return x.replace(' ', '').split('B')[-1]
    
```

```python
df['Memory_Type'] = df['Memory'].apply(convert_to_type)
df.head()
df['Memory_Type'].value_counts()
```

```python
df.drop(columns = ['Memory', 'ScreenResolution'], inplace = True)
df.head()
```

```python
df['SSD'] = 0.2
df.head(2)
```

```python
df['SSD'] = df['Memory_Type'].apply(lambda x: 1 if x == 'SSD' else 0)
df.head()
```

```python
df['Weight'] = df['Weight'].apply(lambda x: x[:-2] if x.endswith('kg') else x)
df.head()
```

```python
df['Weight'].describe()
```

```python
df['Weight'] = df['Weight'].astype(float)
df['Weight'].describe()
```

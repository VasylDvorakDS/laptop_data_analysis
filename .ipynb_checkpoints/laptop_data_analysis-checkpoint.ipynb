```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv', encoding='latin1')
df.head()
```

```python
# df['Cpu_Company'] = df['Cpu'].apply(lambda i : i.split(' ')[0])
df['Cpu_Company'] = df['Cpu'].str.split(' ').str.get(0)
df.head()
```

```python
def convert_to_gb(x):
    memory=x.split(' ')[0]
    if memory.endswith('GB'):
        return int(memory.replace('GB', ''))
    else: 
        return int(float(memory.replace('TB', ''))*1024)
```

```python
df['Memory_Amount'] = df['Memory'].apply(convert_to_gb)
#df['Memory_Amount'] = df['Memory'].apply(lambda i: int(i.split(' ')[0].replace('GB', '')) if 'GB' in i else int(float(i.split(' ')[0].replace('TB', ''))*1024))

df.head()
```

```python
def convert_to_type(x):
    
    return x.replace(' ', '').split('B')[-1]
    
```

```python
df['Memory_Type'] = df['Memory'].apply(convert_to_type)
df.head()
df['Memory_Type'].value_counts()
```

```python
df.drop(columns = ['Memory', 'ScreenResolution'], inplace = True)
df.head()
```

```python
df['SSD'] = 0.2
df.head(2)
```

```python
df['SSD'] = df['Memory_Type'].apply(lambda x: 1 if x == 'SSD' else 0)
df.head()
```

```python
df['Weight'] = df['Weight'].apply(lambda x: x[:-2] if x.endswith('kg') else x)
df.head()
```

```python
df['Weight'].describe()
```

```python
df['Weight'] = df['Weight'].astype(float)
df['Weight'].describe()
```

```python
ldata.head()
```

```python
import numpy as np

data = pd.crosstab(index=ldata['Company'], columns=ldata['Cpu_Company'])
data = data.drop(columns=['Samsung']).reset_index()
data



```

```python
# Строим столбчатую многорядную диаграмму
plt.figure(figsize=(8, 6))
# определяем кол-во делений
n_ticks = np.arange(len(data['AMD']))
# определяем сдвиг
offset = 0.2
# определяем ширину столбцов
w = 0.4
# добавляем сдвиг к кол-ву делений
plt.bar(n_ticks - offset, data['AMD'], width=w)
plt.bar(n_ticks + offset, data['Intel'], width=w)
plt.title('Notebooks')
plt.xlabel('Notebooks_manufacture')
plt.ylabel('Count')
plt.legend(['AMD', 'Intel'])
plt.xticks(n_ticks, data['Company'], rotation=50); 
```

3.2 Построим график в относительных величина

```python
# Строим столбчатую многорядную сложенную диаграмму
data = pd.crosstab(index=ldata['Company'], columns=ldata['Cpu_Company'], normalize='index')
data = data.drop(columns=['Samsung']).reset_index()

plt.figure(figsize=(8, 6))
plt.bar(data['Company'], data['AMD'])
plt.bar(data['Company'], data['Intel'], bottom=data['AMD'])
plt.title('Notebooks')
plt.xlabel('Notebooks_manufacture')
plt.ylabel('Fraction notebooks')
plt.legend(['AMD', 'Intel'])
plt.xticks(rotation=50);
```

```python
import seaborn as sns
sns.boxplot(x=ldata['Price_euros'], y=ldata['Cpu_Company'], whis=1.5);
plt.xlabel('Price_euros')
plt.ylabel('Cpu_Company')
plt.title('Distribution of Price_euros by Cpu_Company')
plt.show()

sns.boxplot(x=ldata['Price_euros'], y=ldata['Memory_Type'], whis=1.5)
plt.xlabel('Price_euros')
plt.ylabel('Memory_Type')
plt.title('Distribution of Price_euros by Memory_Type')
plt.show()


sns.boxplot(x=ldata['Price_euros'], y=ldata['Ram'], whis=1.5)
plt.xlabel('Price_euros')
plt.ylabel('Ram')
plt.title('Distribution of Price_euros by Ram')
plt.show()

sns.boxplot(x=ldata['Price_euros'], y=ldata['Company'], whis=1.5)
plt.xlabel('Price_euros')
plt.ylabel('Company')
plt.title('Distribution of Price_euros by Company')
plt.show()


```
